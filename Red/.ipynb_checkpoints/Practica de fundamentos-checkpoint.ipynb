### 2. Imagenes como arreglos

Una imagen no es mas que un arreglo estandar de Numpy que contiene pixeles de puntos de datos. Cuanto mayor sea el numero de pixeles en una imagen mejor sera en su resolucion. Puede que los pixeles soon pequeños bloques de informacion dispuestos en forma de una cuadrilla de 2 D y la productividad  de un pixel se refiere a la informacion de color presente en ella. Para ser procesado por una computadora, una imagen debe convertirse en una forma binaria. El color de una imagen se puede calcular de la siguiente manera.

Numero de colores/sombras = 2^ bpp donde representa bits por pixel.

Naturalmente, mas cantidad de bits/ pixeles, mas colores posibles en las imagenes. La siguiente tabla muestra la relacion mas claramente.

```python
imgen_tabla = mpimg.imread('tabla.png')
imgplot = plt.imshow(imgen_tabla)
plt.show()
```

### Veamos ahora la representacion de los diferentes tipos de imagenes:

### 1. imagen binaria

Una imagen binaria consta de 1 bit/pixel y por lo tanto, solo puede tener ods colores posibles, es decir, blanco o negro. El negro esta representado por el valor o mientras que el 1 representa el blanco.

```python
imgen_tabla=mpimg.imread('imagen_binaria.png')
imgplot=plt.imshow(imgen_tabla)
plt.show()
```

### 2. Imagen en escala de grises

Una imagen en escala de grises consta de 8 bits por pixel. Esto significa que puede tener tonos 256 tonos diferentes donde 0 pixeles representara el color negro mientras 255 denota el blanco. Por ejemplo, la imagen a continuacion muestra en escala de grises representa en forma de arreglo. Una imagen en escal de grises tiene solo 1 canal donde el canal representa la dimencion.

```python
imgen_tabla=mpimg.imread('imagen_escala_gris.png')
imgplot=plt.imshow(imgen_tabla)
plt.show()
```

### 3. imagen coloreada

Las imagenes en color se representan como una combinacion de rojo, azul y verde, y todos los demas colores se pueden lograr mezclando estos colores primarios en las propociciones correctas.

```python
imgen_tabla=mpimg.imread('imagen_coloreada.png')
imgplot=plt.imshow(imgen_tabla)
plt.show()
```

Una imagen en color tambien consta de 8 bits por pixel. Como resultado, se pueden representar 256 tonos de colores diferentes con 0 que detona negro y 255 blanco. Veamos la famosa imagen en color de un mandril que se ha citado en muchos ejemplos de procesamiento de imagen.

```python
imgen_tabla=mpimg.imread('mandril_coloreado.jpg')
implot=plt.imshow(imgen_tabla)
plt.show()
```

i compobamos la forma de la imagen de arriba, obtendriamos:

Forma (288,288,3)288:ancho de pixel, 288:altura de pixel, 3:canal de color

Esto significa que podemos representar la imagen de arriba en forma de un arreglo tridimencional.

Lea la imagen con la funcion Imread. Utilizaremos la imagen coloreada "mandril" para fines de demostracion.

```python
img_raw= cv2.imread('mandril_color.png')
```

El tipo y la forma del arreglo

Por lo tanto, la imagen.png se transformaba en un arreglo numpy con una forma de 288x288 y tine 3 canales.

viendo la imagen.

```python
plt.imshow(img_raw)
```

Lo que obtenemos como resultado es un poco diferente con respecto al color. Esperabamos con la imagen de color brillantes, pero lo que obtenemos es una imagen con cierto matiz azulado. Eso sucede porque OpenCV y matplolib tiene diferentes ordenes de colores primarios. Mientras que OpenCV lee imagenes en forma de BGR, matplolib, por otro lado, sigue el orden de RGB. Por lo tanto , cuando leemos un archivo a traves de OpenCV, lo leemos como si contuviera canales en el oreden de azul, verde, rojo. Sin embargo, cuando mostramos la imagen usando matplotlib, el canal rojo y el azul se intercambian y, por lo tanto, el tinte azul. Para evitar este problema, transformaremos el canal a como matplotlib espera que use la funcion cvtColor.


```python
img_rgb=cv2.cvtColor(img_raw,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

### Operaciones basicas en imagenes 

En esta seccion, aprenderemos como podemos dibujar varias formas en una imagen existente para tener una idea de como trabajar con OpenCV.

### Dibujando en imagenes
.
Crea una imagen en negro que actuara como plantilla

```python
image_blank=np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
# Muestra la imagen en negro
plt.imshow(image_blank)
```

### Funciones  y atributos

La funcion general para dibujar formas o figuras en la imagen es:

#### cv2.shape(line,rectangle etc)(image,Pt1,Pt2,color,thickness)

Hay algunos argumentos comunes que se pasan en funcion para dibujar figuras en las imagenes:

1)Imagen en la que se dibujaran formas 

2)Coordenadas de la figuara a dibujar desde Pl1(arriba a la izquierda) a Pl2(abajo a la derecha)

3)Color; el color de la dorma que se va a dibujar requiere especificar los puntos, a traves de los cuales pasara la linea.

4)El grosor de la figura geometrica.

### a) Linea recta

Dibujar una linea recta a traves de una imagen requiere especificar los puntos, a traves de los cuales pasara la linea.

```python
# Dibujar una linea roja diagonal con un grosor de 5 px
line_red = cv2.line(image_blank,(0,0),(511,511),(255,0,0),5)
plt.imshow(line_red)
```

```python
# Dibujar una linea verde diagonal con un grosor de 5 px
line_green = cv2.line(image_blank,(0,0),(511,511),(0,255,0),5)
plt.imshow(line_red)
```

### b) Rectangulo

Para dibujar un rectangulo necesitamos especificar las coordenadas superio izquierda e inferios derecha.

```python
# Dibujar un rectangulo azul con un grosor de 5 px
rectangle= cv2.rectangle(image_blank,(384,0),(510,128),(0,0,255),5)
plt.imshow(rectangle)
```

### c) Escribir en imagenes

Agregar texto a las imagenes tambien es similar a dibujar formas en ellas. Pero debe especificar ciertos argumentos antes de hacerlo:

1)Texto a escribir

2)Coordenadas del texto. El texto en una imagen comienza desde la direccion inferior izquierda.

3)Tipos de fuente escala.

4)Otros atributos como color, grosor y tipo de linea. Normalmente, el tipo de linea que se usa es lineType=cv2.LINE_AA


```python
font=cv2.FONT_HERSHEY_SIMPLEX
text=cv2.putText(image_blank,'OpenCV',(10,300), font, 4, (255,255,255),2,cv2.LINE_AA)
plt.imshow(text)
```

Estas fueron las operaciones basicas que se pueden hacer en imagenes usando OpenCV. Sientate libre de experimentar con las figuras y el texto.
