### 2. Imagenes como arreglos

Una imagen no es mas que un arreglo estandar de Numpy que contiene pixeles de puntos de datos. Cuanto mayor sea el numero de pixeles en una imagen mejor sera en su resolucion. Puede que los pixeles soon pequeños bloques de informacion dispuestos en forma de una cuadrilla de 2 D y la productividad  de un pixel se refiere a la informacion de color presente en ella. Para ser procesado por una computadora, una imagen debe convertirse en una forma binaria. El color de una imagen se puede calcular de la siguiente manera.

Numero de colores/sombras = 2^ bpp donde representa bits por pixel.

Naturalmente, mas cantidad de bits/ pixeles, mas colores posibles en las imagenes. La siguiente tabla muestra la relacion mas claramente.

```python
imgen_tabla = mpimg.imread('tabla.png')
imgplot = plt.imshow(imgen_tabla)
plt.show()
```

### Veamos ahora la representacion de los diferentes tipos de imagenes:

### 1. imagen binaria

Una imagen binaria consta de 1 bit/pixel y por lo tanto, solo puede tener ods colores posibles, es decir, blanco o negro. El negro esta representado por el valor o mientras que el 1 representa el blanco.

```python
imgen_tabla=mpimg.imread('imagen_binaria.png')
imgplot=plt.imshow(imgen_tabla)
plt.show()
```

### 2. Imagen en escala de grises

Una imagen en escala de grises consta de 8 bits por pixel. Esto significa que puede tener tonos 256 tonos diferentes donde 0 pixeles representara el color negro mientras 255 denota el blanco. Por ejemplo, la imagen a continuacion muestra en escala de grises representa en forma de arreglo. Una imagen en escal de grises tiene solo 1 canal donde el canal representa la dimencion.

```python
imgen_tabla=mpimg.imread('imagen_escala_gris.png')
imgplot=plt.imshow(imgen_tabla)
plt.show()
```

### 3. imagen coloreada

Las imagenes en color se representan como una combinacion de rojo, azul y verde, y todos los demas colores se pueden lograr mezclando estos colores primarios en las propociciones correctas.

```python
imgen_tabla=mpimg.imread('imagen_coloreada.png')
imgplot=plt.imshow(imgen_tabla)
plt.show()
```

Una imagen en color tambien consta de 8 bits por pixel. Como resultado, se pueden representar 256 tonos de colores diferentes con 0 que detona negro y 255 blanco. Veamos la famosa imagen en color de un mandril que se ha citado en muchos ejemplos de procesamiento de imagen.

```python
imgen_tabla=mpimg.imread('mandril_coloreado.jpg')
implot=plt.imshow(imgen_tabla)
plt.show()
```
